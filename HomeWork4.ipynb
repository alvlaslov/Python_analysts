## Задание 1

### 1.1. Скачать данные по ссылке

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('kc_house_data.csv', sep=',', encoding ='cp1251', quoting=3)
df.info()
```

```python
df.head()
```

```python
# уберем лишние кавычки из файлы с данными и переведем формат данных float64

df['date'] = df['date'].str.replace('""', '')
df['zipcode'] = df['zipcode'].str.replace('""', '')
df['floors'] = df['floors'].str.replace('""', '')
df['id'] = df['id'].str.replace('"', '')
df['sqft_lot15'] = df['sqft_lot15'].str.replace('"', '')

df['zipcode'] = df['zipcode'].astype('int64')
df['floors'] = df['floors'].astype('float64')
df['id'] = df['id'].astype('int64')
df['sqft_lot15'] = df['sqft_lot15'].astype('float64')

df.head()
```

### 1.2 Изучите стоимости недвижимости

Выводы:

Из графического отображения данных видно, что:
- Большая часть домов имеет стоимость от 75 000 до 700 000
- Меньшая часть домов имеет стоимость свыше 700 000

Выводы:

## Задание 3


Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.

Построим матрицу корреляций и определим какие характеристики влияют на стоимость недвижимости.

```python
df['waterfront'].replace({'no': 0, 'yes': 1}, inplace = True)
corr_matrix = df.corr()
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
corr_matrix
```

```python
plt.figure(figsize=(8, 8))
sns.heatmap(corr_matrix, cmap='coolwarm', annot = True, linewidths = .5);
```

Согласно матрице корреляций, наибольшие характеритсики, влияющие на стоимость дома, являются: sqft_living, grade, sqft_above, sqft_living15, bathrooms (прямая линейная связь).

Для анализа подойдут графики взаимного распределения. К примеру, хотим проанализировать зависимость жилой площади и стоимость недвижимости. У нас есть гипотеза, что чем больше жилая площадь дома, тем выше его цена.

```python
sns.jointplot(x=df['sqft_living'], y=df['price'], kind='reg');
```

Есть неплохая связь: чем больше жилая площадь дома, тем выше его цена, так что гипотеза подтверждается. 

Хотим проанализировать стоимость недвижимости в зависимости от вида на набережную. Используем графики боксплот.

```python
df['waterfront'].replace({0: 'no', 1: 'yes'}, inplace = True)
plt.figure(figsize=(16, 8))
sns.boxplot(x=df['price'], y=df['waterfront'], whis=1.5, showfliers=False) # без выбросов
plt.title('Стоимость недвижимости в зависимости от вида на набережную')
plt.xlabel('price')
plt.ylabel('waterfront');
```

Согласно графику видим, что стоимость недвижимости с видом на набережную в среднем выше стоимости недвижимости без вида на набережную.

У нас есть гипотеза, что чем выше оценка дома, тем выше его цена.

```python
sns.jointplot(x=df['price'], y=df['grade'], kind='reg');
```

Есть связь: чем больше чем выше оценка дома,, тем выше его цена, так что гипотеза подтверждается.

У нас есть гипотеза, что чем больше количество ванных комнат, тем выше его цена.


```python
sns.jointplot(x=df['price'], y=df['bathrooms'], kind='reg');
```

Есть связь: чем больше количество ванных комнат, тем выше его цена, так что гипотеза подтверждается.
